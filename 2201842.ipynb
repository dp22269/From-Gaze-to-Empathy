```python
#feature selection
import pandas as pd
import os

# path
csv_directory = 'EyeT/EyeT/'

# Create an empty list to hold the resulting dataframes
dataframe_list = [] 

# Loop through each file in the directory
for filename in os.listdir(csv_directory):
    # Check if the file is a CSV file
    if filename.endswith('.csv'):
        # Read the CSV file 
        current_dataframe = pd.read_csv(os.path.join(csv_directory, filename))
        
        numeric_columns = current_dataframe.select_dtypes(include=['float64', 'int64'])#Select columns
        mean_values = numeric_columns.mean().to_frame().T.iloc[:, 0:]#take a mean and perform traanspose
        first_row = pd.read_csv(os.path.join(csv_directory, filename), usecols=["Participant name"]).head(1)#add participant name
        concatenated_dataframe = pd.concat([mean_values, first_row], axis=1)#concat all 502 files
        
        dataframe_list.append(concatenated_dataframe)# Append the resulting dataframe to the list


final_df = pd.concat(dataframe_list)# Concatenate all the resulting data into a single data

```

```python
# Define the input and target variables
X = df[['Recording timestamp', 'Gaze point X', 'Gaze point Y','Gaze event duration','Eye movement type index','Fixation point X', 'Fixation point Y']]
y = df['Total Score original']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Test the model on the testing set
predictions = model.predict(X_test)

# Calculate the R^2 score of the model
r_squared = model.score(X_test, y_test)
print("R^2 score:", r_squared)
```
